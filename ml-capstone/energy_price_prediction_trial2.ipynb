### Building a new Feature

We've seen detailed features for energy generation with each feature representing a source like biomass, hydropower, nuclear etc. At the same time we have a single feature for total consumption. This kind of feature structure complicates our ML Model due to the increase in the number of features for generation where as we have a single feature for total consumption.<br>
Unless we're interested to know how the different energy sources are correlated with one another, this detail of information does not help us.<br>
If we reduce the actual generation features to a single **actual generation[mwh]** feature, we could directly find a correlation between generation and consumption which in other words represents supply and demand for the energy price prediction.

```python
# build new feature "actual generation[mwh]"

master_data['actual_generation[mwh]'] = (master_data['biomass[mwh]'] + master_data['hydropower[mwh]'] + master_data['wind_offshore[mwh]'] + 
                                        master_data['wind_onshore[mwh]'] + master_data['photovoltaics[mwh]'] + master_data['other_renewable[mwh]'] +  
                                        master_data['nuclear[mwh]'] + master_data['fossil_brown_coal[mwh]'] + master_data['fossil_hard_coal[mwh]'] +  
                                        master_data['fossil_gas[mwh]'] + master_data['hydro_pumped_storage[mwh]'] + master_data['other_conventional[mwh]'])
```
